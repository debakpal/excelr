```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df = df.rename({'User.ID':'UserId','Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
df
```

```python
books_new = df.pivot_table(index='UserId', columns='Title', values='Rating')
books_new
```

```python
books_new.fillna(0, inplace=True)
```

```python
books_new.index = df['UserId'].unique()
books_new
```

```python
user1 = 1 - pairwise_distances( books_new.values, metric='cosine')
user1
```

```python
user = pd.DataFrame(user1)
user
```

```python
user.index = df.UserId.unique()
user.columns = df.UserId.unique()
```

```python
def recommend(custID):
    simID = user.loc[custID].idxmax()
    simID_df = df[df['UserId'] == simID].Title
    custID_df = df[df['UserId'] == custID].Title
    return set(simID_df) - set(custID_df)
```

```python
recommend(276813)
```
